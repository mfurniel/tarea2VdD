```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
libro = open('libro.txt', 'r').read()
```

```python
stopwords = set(STOPWORDS)
```

```python
meta = WordCloud(
    background_color='white',
    stopwords=stopwords
)
meta.generate(libro)
```

```python
from PIL import Image
mascara = np.array(Image.open('ejem.png'))
```

```python
from stop_words import get_stop_words
no_palabras= get_stop_words('spanish')

meta = WordCloud(background_color='black', color_func=lambda *args, **kwargs: "Violet", mask=mascara, stopwords=no_palabras)

# generamos la nube
meta.generate(libro)

# la desplegamos
fig = plt.figure()
fig.set_figwidth(15)
fig.set_figheight(15)

plt.imshow(meta, interpolation='lanczos')
plt.axis('off')
plt.show()
```
