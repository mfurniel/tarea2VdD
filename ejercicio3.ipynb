```python
import pandas as pd
import plotly.express as px

data = pd.read_csv('datasets/gapminderData2.csv')
```

```python
fig = px.scatter(
    data, x="gdpPercap", 
    y="bornPerwom", 
    animation_frame="year",
    size="pop", 
    color="continent", 
    hover_name="country",
    log_x=True, 
    size_max=100, 
    range_x=[200,55000],
    title="Relación entre el PIB per Cápita y los Nacimientos por Mujer",
)

fig.update_layout( 
    xaxis_title="Nacimientos por Mujer",
    yaxis_title="PIB per Cápita",
    annotations=[
        dict(
            text="Este gráfico muestra la relación entre el PIB per Cápita y los Nacimientos por Mujer en relacion al tamaño de poblacion a nivel mundial a lo largo de los años.",
            align="left",
            showarrow=False,
            xref="paper",
            yref="paper",
            x=-0.01,
            y=1.1,
            # font=dict(
            #     size=12
            # )   
        )
    ],
    width=1200,  # Ancho de la figura en píxeles
    height=600
)

fig.show()
```
